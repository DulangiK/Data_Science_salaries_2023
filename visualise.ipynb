```python
import pandas as pd
import seaborn as sns
```

### Reading the csv file

```python
data = pd.read_csv("ds_salaries.csv")
```

### How many job positions are discussed in the dataset?

download.jpeg







```python
n_job_title = len(pd.unique(data['job_title']))
  
print("No.of.job positions discussed in this dataset :", n_job_title)
```

```python
data.job_title.unique()
```

### Filter out job positions containing the key word "data" :

```python
filtered_rows = data[data['job_title'].str.contains('Data')]
filtered_rows
```

### Filter out data scientist job positions

```python
data_sci=filtered_rows[filtered_rows['job_title'].str.contains('Scientist')]
data_sci
```

### Filter out the data scientist job positions based on US companies

```python
data_sci_us=data_sci[data_sci['company_location'].str.contains('US')]
data_sci_us
```

### The salary range for data scientist working in US based companies

```python
min_salary_us_data_scientist = data_sci_us["salary"].min()
max_salary_us_data_scientist = data_sci_us["salary"].max()
print("The salary range of a data scientist working in US company",min_salary_us_data_scientist,":",max_salary_us_data_scientist)
```

### The work experience of data scientist working in US based companies 

```python
data.experience_level.unique()
```

### Filter out based on SE work experience category

```python
data_sci_us_SE=data_sci_us[data_sci_us['experience_level'].str.contains('SE')]
data_sci_us_SE
```

### salary ranges based on SE work experience for data scientists working in US based companies

```python
sns.histplot(data=data_sci_us_SE, x="salary", kde=True , color="purple")
```

```python
data_sci_us_SE['salary'].describe()
```
